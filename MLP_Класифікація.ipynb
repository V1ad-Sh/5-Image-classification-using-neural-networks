```python
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Текстові представлення символів для 0 і 1
zero = """
.xxx
.x.x
.x.x
.xxx

xxxx
x..x
x..x
xxxx

xxx.
x.x.
x.x.
xxx.

xxx.
x.x.
xxx.
....

.xxx
.x.x
.xxx
....

....
xxx.
x.x.
xxx.

....
.xxx
.x.x
.xxx

.xx.
x..x
x..x
.xx.

..x.
.x.x
.x.x
..x.

.x..
x.x.
x.x.
.x..
"""

one = """
...x
...x
...x
...x

..x.
..x.
..x.
..x.

.x..
.x..
.x..
.x..

x...
x...
x...
x...

..xx
...x
...x
...x

...x
..xx
...x
...x

.xx.
..x.
..x.
..x.

xx..
.x..
.x..
xxx.

.xx.
..x.
..x.
.xxx

.x..
xx..
.x..
.x..
"""

# Створення векторів зображень та міток
n = zero + one
n = n.split('\n\n')
x = []
for i in n:
    s = i.replace('\n', '').replace('.', '0').replace('x', '1')
    x.append(np.array([int(e) for e in s]))
x = np.array(x)
y = np.array([0] * 10 + [1] * 10)  # 10 цифр для 0 та 10 для 1

# Розбивка на навчальні і тестові дані
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Визначаємо діапазон для кількості нейронів у прихованому шарі
param_grid = {
    'hidden_layer_sizes': [(i,) for i in range(1, 21)]  # Перевіряємо кількість нейронів від 1 до 20
}

# Налаштування моделі MLP з GridSearchCV
mlp = MLPClassifier(max_iter=1000, random_state=42)
grid_search = GridSearchCV(mlp, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Найкращі гіперпараметри та точність моделі на тестових даних
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
best_accuracy = accuracy_score(y_test, y_pred)

grid_search.best_params_, best_accuracy
```
